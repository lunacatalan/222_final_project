````python
import pystac_client
import planetary_computer
import odc.stac
import matplotlib.pyplot as plt

from pystac.extensions.eo import EOExtension as eo
````

````python
catalog = pystac_client.Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)
````

# Select data from 1984-2021

````python
# Define your temporal range
daterange = {"interval": ["1984-03-22T00:00:00Z", "1984-12-22T00:00:00Z"]}

# Define your search with CQL2 syntax
search_1984 = catalog.search(filter_lang="cql2-json", filter={
  "op": "and",
  "args": [
    {"op": "anyinteracts", "args": [{"property": "datetime"}, daterange]}, # date
    {"op": "=", "args": [{"property": "collection"}, "landsat-c2-l2"]}, # from which satellite
    {"op": "=", "args": [{"property": "landsat:wrs_path"}, "165"]}, # path found from exploring landsat data
    {"op": "=", "args": [{"property": "landsat:wrs_row"}, "035"]}, # row found from exploring landsat data
    {"op": "<=", "args": [{"property": "eo:cloud_cover"}, 50]} # less than 50% cloud cover
  ]
})

item_84 = search_1984.item_collection()
search_84 = item_84[1] # image from Sept, 23, 1984
````

# Make a function to calculate and store the NDVI means for each year 

````python
def ndvi_mean(search):
    bands_of_interest = ["nir08", "red", "green", "blue"]
    
    data = odc.stac.stac_load(
        [search], 
        bands=bands_of_interest
    ).isel(time=0)
    
    # choose the bands from the data
    red = data["red"].astype("float")
    nir = data["nir08"].astype("float")
    
    # calculate the ndvi
    ndvi = (nir - red) / (nir + red)
    
    # clip rasters to aoi
    ndvi_buffer = ndvi.rio.clip_box(*radius_gdf.total_bounds)
    
    # get the mean values in the ndvi buffer
    ndvi_mean = ndvi_buffer.values.mean()
    
    return(ndvi_mean)
````
